```python
import pandas as pd
```

```python
weather_data = pd.read_csv("C:/Users/sonal/flight-delay-prediction/dataset/JFK-weather-GMT.csv")
```

```python
weather_data.info()
```

```python
#Renaming all columns for ease
weather_data = weather_data.rename(columns={
    'time' : 'time',
    'temperature_2m (°C)' : 'temperature',
    'relative_humidity_2m (%)' : 'humidity',
    'dew_point_2m (°C)' : 'dew_point',
    'precipitation (mm)' : 'precipitation',
    'rain (mm)' : 'rain',
    'snowfall (cm)' : 'snowfall',
    'weather_code (wmo code)' : 'weather_code',
    'pressure_msl (hPa)' : 'pressure',
    'surface_pressure (hPa)' : 'surface_pressure',
    'cloud_cover (%)' : 'cloud_cover',
    'wind_speed_10m (km/h)' : 'wind_speed_10m',
    'wind_speed_100m (km/h)' : 'wind_speed_100m',
    'wind_direction_10m (°)' :'wind_direction_10m',
    'wind_direction_100m (°)' : 'wind_direction_100m',
    'wind_gusts_10m (km/h)' : 'wind_gusts_10m',
    'is_day ()' : 'is_day' 
})
```

```python
#Checking for duplicates, none found
print("Number of duplicate rows: ", weather_data.duplicated().sum())
```

```python
#get unique values from each column to correct data before pre-processing
unique_values_dict = {}
for column in weather_data.columns:
    unique_values_dict[column] = weather_data[column].unique()

# Print unique values for each column
for column, unique_values in unique_values_dict.items():
    print(f"Unique values in '{column}': {unique_values}")
```

```python
#Rechecking null counts after dropping irrelevant reports
null_counts = weather_data.isnull().sum()
null_counts = null_counts[null_counts > 0]  # Filter to only columns with missing values
print(null_counts)
```

```python
weather_data['time'] = pd.to_datetime(weather_data['time'])
weather_data['mod_time'] = weather_data['time'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
#Saving df to csv for merging
weather_data.to_csv('C:/Users/sonal/flight-delay-prediction/dataset/weather_cleaned.csv',index=False)
```
